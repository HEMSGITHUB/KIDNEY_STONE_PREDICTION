### Distribution Of Calcium Concentration In The Tested People

```python
p=sns.factorplot('calc',data=data,kind='count',hue='target')
p.set_xlabels('Calcium')
p.set_ylabels('')
p.fig.set_figheight(3)
p.fig.set_figwidth(100)
print('Here on y axis we have calcium concentration in urine with target differentiation.')
print('We can observe that maximum affected group are lying at high calcium Concentrations.')
```

### Specific Gravity Relation With Kidney Stones Formation

```python
p=sns.factorplot('gravity',data=data,kind='count',hue='target')
p.set_xlabels('Specific Gravity')
p.set_ylabels('')
p.fig.set_figheight(3)
p.fig.set_figwidth(100)
print('Here on y axis we have Specific Gravity of the urine with target differentiation.')
print('We can observe that increase in specific gravity can cause increased risk of stones formation in the kidneys.')
```

```python
plt.figure(figsize=(10,5))
plt.scatter(data['calc'],data['gravity']) 
plt.title('Relation Between Specific gravity and Calcium conc')
plt.xlabel('Calcium Concentration')
plt.ylabel('Specific Gravity')
plt.show()
```

### Urea Concentration And Specific Gravity Relation

```python
plt.figure(figsize=(10,5))
plt.scatter(data['urea'],data['osmo']) 
plt.title('Relation Between Specific gravity and Calcium conc')
plt.xlabel('Urea Concentration ')
plt.ylabel('Osmolarity')
plt.show()
print('We can see that osmolarity of the urine is gradually increased with increase in urea concentration.')
```

```python
data["osmo"].hist(figsize=(15,4), color = 'green')
plt.title("Osmolarity Distribution")
plt.xlabel('Osmolarity')
plt.show()
```

```python
#Propagation by variable
def PropByVar(df, variable):
    dataframe_pie = df[variable].value_counts()
    ax = dataframe_pie.plot.pie(figsize=(7,7), autopct='%1.2f%%', fontsize = 12)
    ax.set_title(variable + ' \n', fontsize = 15)
    return np.round(dataframe_pie/df.shape[0]*100,2)
PropByVar(data, 'target')
```
